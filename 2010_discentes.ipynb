```python
import pandas as pd
discentes_2010 = pd.read_csv('discentes-2010.csv', sep=';')
discentes_2010.head()
```

```python
cet2010 = discentes_2010[(discentes_2010.id_curso == 10320810) & (discentes_2010.periodo_ingresso == 2)]
```

```python
cet2010.head()
```

```python
statusCET_2 = cet2010['status'].value_counts()
```

```python
# import numpy as np
import matplotlib.pyplot as plt
```

```python
dfCET = pd.DataFrame(statusCET_2)

dfCET.head()
```

```python
plt.figure(figsize=(3,3))

plt.pie(x = dfCET['status'], labels=dfCET.index, shadow=False, startangle=90, autopct= '%1.1f%%')
plt.title('Status dos discentes de Ciências e Tecnologia 2010.2 atualmente(Março/2022)')

plt.show()
```
